```python
drop_cols = ['ARREST_KEY', 'ARREST_DATE', 'PD_CD', 'PD_DESC', 'KY_CD','LAW_CODE', 'LAW_CAT_CD','ARREST_PRECINCT',
       'JURISDICTION_CODE', 'AGE_GROUP','PERP_RACE', 'X_COORD_CD',
       'Y_COORD_CD', 'Latitude', 'Longitude', 'New Georeferenced Column' ]

df.drop(drop_cols, inplace = True, axis=1)
```

```python
df.head(60)
```

```python
df_filtered = df[df['OFNS_DESC'] == 'RAPE']
```

```python
print(df_filtered)
```

```python
df_filtered.head(50)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='ARREST_BORO', data=df_filtered, order=df_filtered['ARREST_BORO'].value_counts().index)
plt.title('Number of Rapes in Each Borough')
plt.xlabel('Borough')
plt.ylabel('Number of Rapes')
plt.show()
```

In conclusion the Bronx has the highest rape year to date. Staten Island has the Least amount of Rapes.
